```python
import numpy as np
import pandas as pd
```

```python
dataframe_6 = pd.read_csv('data_ria_news_v2.6_from_nov_to_sept.csv', delimiter=',')
dataframe_7 = pd.read_csv('data_ria_news_v2.7_from_sept.csv', delimiter=',')
dataframe_8 = pd.read_csv('data_ria_news_v2.8_from_sept.csv', delimiter=',')
dataframe_9 = pd.read_csv('data_ria_news_v2.9_from_aug.csv', delimiter=',')
dataframe_10 = pd.read_csv('data_ria_news_v2.10_from_aug.csv', delimiter=',')
```

```python
dataframe_8
```

```python
dataframe_merged67 = pd.merge(dataframe_6, dataframe_7,  how="outer")
```

```python
dataframe_merged67
```

```python
dataframe_merged678 = pd.merge(dataframe_merged67, dataframe_8,  how="outer")
dataframe_merged6789 = pd.merge(dataframe_merged678, dataframe_9,  how="outer")
dataframe_merged_all = pd.merge(dataframe_merged6789, dataframe_10,  how="outer")

dataframe_merged_all
```

```python
dataframe_merged_all[dataframe_merged_all.duplicated(subset=['Date', 'Text'], keep="first")]
```

```python
df_del_dubl = dataframe_merged_all.drop_duplicates(subset=['Date', 'Text'], keep="first", inplace=False)
df_del_dubl
```

```python
df_del_dubl.to_csv("clean_ria_data.csv")
```
